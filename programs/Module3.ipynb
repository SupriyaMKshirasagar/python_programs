```python
from math import sqrt
x=[1,2,3,4,5,6]
N=len(x)
mu=sum(x)/N
x_mu=0
for i in x:
    x_mu=x_mu+(i-mu)**2
print("x_mu:",x_mu)
    
sigma=sqrt((x_mu)/N)  
print("Standard Deviation:",sigma)  
```

```python
import numpy as np
import matplotlib.pyplot as plt

vec=np.array([[-3],[2]])
origin=np.zeros(vec.shape)

plt.figure(figsize=(6,6))
plt.quiver(*origin,*vec,color=['b'],scale=1,units='xy')

plt.grid()
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.gca().set_aspect('equal')
plt.show()
```
