```python
from math import sqrt
x=[-2,-1,1,4,6,8]
N=len(x)
mu=sum(x)/N
x_mu=0
for i in x:
    x_mu=x_mu+(i-mu)**2
print("x_mu:",x_mu)
    
sigma=sqrt((x_mu)/N)  
print("Standard Deviation:",sigma)  
```
